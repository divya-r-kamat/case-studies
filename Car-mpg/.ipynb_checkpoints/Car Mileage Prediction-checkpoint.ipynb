#### Create the low, mid, high value for mpg column and the range is; 1.  min value to 17 = low, 2. 17 to 29 = medium and 3. 29 to the max value is high. 

```python
def mpg_level(df_row):
    try:
        if df_row < 17:
            mpg_level = 'low'
        elif df_row >= 17 and df_row < 29:
            mpg_level = 'medium'
        else:
            mpg_level = 'high'
    except:
        pass
    return mpg_level
        
        
```

```python
df_car['mpg_level'] = df_car['mpg'].apply(mpg_level)
```

```python
from sklearn.model_selection import train_test_split 
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
```

```python
columns = ['horsepower','acceleration','displacement','cylinders','model_year','weight']
X = df_car[columns]
y= df_car.mpg
```

```python
X = sm.add_constant(X)
```

```python
# create training and testing vars
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
mod = sm.OLS(y_train,X_train)
res = mod.fit()
```

```python
print_model = res.summary()
print(print_model)
```

```python
res.params
```

```python
# fit a model
lm = LinearRegression()
model = lm.fit(X_train, y_train)
predictions = lm.predict(X_test)
```

```python
lm.intercept_
```

```python
lm.coef_
```

```python
#plt.scatter(y_test,X_test)
plt.scatter(y_test, predictions)
plt.xlabel('True Values')
plt.ylabel('Predictions')
```
